```python
import itertools

import pandas as pd

import bib_parse
```

```python
cited = bib_parse.bib_to_df('../bibtex/2017-02-01-times_cited_descending.bib')
newest = bib_parse.bib_to_df('../bibtex/2017-02-01-publication_date_newest.bib')
oldest = bib_parse.bib_to_df('../bibtex/2017-02-01-publication_date_oldest.bib')
c180 = bib_parse.bib_to_df('../bibtex/2017-02-01-useage_count_180.bib')
c2013 = bib_parse.bib_to_df('../bibtex/2017-02-01-useage_count_2013.bib')

dfs = [cited, newest, oldest, c180, c2013]
```
